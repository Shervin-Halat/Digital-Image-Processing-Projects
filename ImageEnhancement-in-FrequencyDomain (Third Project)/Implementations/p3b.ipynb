```python
import numpy as np
from math import e
from PIL import Image
import glob
import cv2
from matplotlib import pyplot as plt
```

```python
images = [cv2.imread(file,0) for file in glob.glob("C:/Users/sherw/OneDrive/Desktop/DIP_HW3/inputs/p3/*.png")]
```

```python
index = 0
for im in images:
    #matplotlib.rcParams['figure.figsize'] = [3,3]
    plt.figure(index)
    index+=1
    dft = np.fft.fft2(im)
    dft2 = np.fft.fftshift(dft)
    angle = np.angle(dft2)
    mag = np.abs(dft2)  
    plt.subplot(1,2,1)
    plt.imshow(np.log(mag),'gray')
    plt.title('magnitude')
    plt.subplot(1,2,2)
    plt.imshow(angle, 'gray')
    plt.title('phase')
    plt.suptitle('image {}'.format(index))
    plt.show()
```
