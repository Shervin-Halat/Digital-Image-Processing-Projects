```python
def pix_med_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = median(neigh_pixs)
    return out_pixel
def median_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_med_filter([i,j], image, filter_size)
    fil_size = filter_size
    return filtered_image
```

```python
def pix_midpoint_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = 1/2*(max(neigh_pixs)+min(neigh_pixs))
    return out_pixel
def midpoint_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_midpoint_filter([i,j], image, filter_size)
    fil_size = filter_size
    return filtered_image
```

```python
def pix_max_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = max(neigh_pixs)
    return out_pixel
def max_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_max_filter([i,j], image, filter_size)
    fil_size = filter_size
    return filtered_image
```

```python
def pix_min_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = min(neigh_pixs)
    return out_pixel
def min_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_min_filter([i,j], image, filter_size)
    fil_size = filter_size
    return filtered_image
```

```python
def pix_arithmetic_mean_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = mean(neigh_pixs)
    return out_pixel
def arithmetic_mean_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_arithmetic_mean_filter([i,j], image, filter_size)
    fil_size = filter_size
    return filtered_image
```

```python
def pix_geometric_mean_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    if 0 in neigh_pixs:
        out_pixel = 0
    else:
        out_pixel = geometric_mean(neigh_pixs)
    return out_pixel
def geometric_mean_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_geometric_mean_filter([i,j], image, filter_size)
    fil_size = filter_size
    return filtered_image
```

```python
def pix_harmonic_mean_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = harmonic_mean(neigh_pixs)
    return out_pixel
def harmonic_mean_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_harmonic_mean_filter([i,j], image, filter_size)
    fil_size = filter_size
    return filtered_image
```

```python
def pix_contraharmonic_mean_filter(pixelcoor, image2, filtershape,Q):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    neigh_pixs = np.array(neigh_pixs).astype(np.float)
    if Q>=0:
        out_pixel = np.sum(neigh_pixs**(Q+1)) / (np.sum(neigh_pixs**(Q))+eps)
    else:
        neigh_pixs += eps
        out_pixel = np.sum(neigh_pixs**(Q+1)) / (np.sum(neigh_pixs**(Q))+eps)
    return out_pixel
def contraharmonic_mean_filter(image,filter_size,Q):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_contraharmonic_mean_filter([i,j], image, filter_size,Q)
    fil_size = filter_size
    return filtered_image
```

```python
def distance(point1,point2):
    return sqrt((point1[0]-point2[0])**2 + (point1[1]-point2[1])**2)

def gaussianHP(D0,image,u,v):
    base = np.ones(image.shape[0:2])
    rows, cols = image.shape[0:2]
    for y in range(cols):
        for x in range(rows):
            base[x,y] = 1-exp(((-distance((x,y),(u,v))**2)/(2*(D0**2))))
    return base

def hor_line(D0, image, u):
    base = np.ones(image.shape[0:2])
    rows, cols = image.shape[0:2]
    for y in range(cols):
        for x in range(rows):
            if abs(x-u) <= D0:
                base[x,y] = 0
    return base

def notch_gauss(image, x, y,D0):
    a,b = image.shape
    filt1 = gaussianHP(D0, image, x, y)
    filt2 = gaussianHP(D0, image, (a/2-x)*2 + x, (b/2-y)*2 + y)
    filt = filt1*filt2
    return filt

def notch_hor(image, x, D0):
    a,b = image.shape
    filt1 = hor_line(D0, image, x)
    filt2 = hor_line(D0, image, (a/2-x)*2 + x)
    filt = filt1*filt2
    return filt
```
