```python
def pix_med_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = median(neigh_pixs)
    return out_pixel
def median_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_med_filter([i,j], image, filter_size)
    fil_size = filter_size
    return(filtered_image, fil_size)
```

```python
def pix_midpoint_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = 1/2*(max(neigh_pixs)+min(neigh_pixs))
    return out_pixel
def midpoint_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_midpoint_filter([i,j], image, filter_size)
    fil_size = filter_size
    return(filtered_image, fil_size)
```

```python
def pix_max_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = max(neigh_pixs)
    return out_pixel
def max_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_max_filter([i,j], image, filter_size)
    fil_size = filter_size
    return(filtered_image, fil_size)
```

```python
def pix_min_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = min(neigh_pixs)
    return out_pixel
def min_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_min_filter([i,j], image, filter_size)
    fil_size = filter_size
    return(filtered_image, fil_size)
```

```python
def pix_arithmetic_mean_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = mean(neigh_pixs)
    return out_pixel
def arithmetic_mean_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_arithmetic_mean_filter([i,j], image, filter_size)
    fil_size = filter_size
    return(filtered_image, fil_size)
```

```python
def pix_geometric_mean_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    if 0 in neigh_pixs:
        out_pixel = 0
    else:
        out_pixel = geometric_mean(neigh_pixs)
    return out_pixel
def geometric_mean_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_geometric_mean_filter([i,j], image, filter_size)
    fil_size = filter_size
    return(filtered_image, fil_size)
```

```python
def pix_harmonic_mean_filter(pixelcoor, image2, filtershape):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    out_pixel = harmonic_mean(neigh_pixs)
    return out_pixel
def harmonic_mean_filter(image,filter_size):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_harmonic_mean_filter([i,j], image, filter_size)
    fil_size = filter_size
    return(filtered_image, fil_size)
```

```python
def pix_contraharmonic_mean_filter(pixelcoor, image2, filtershape,Q):
    size = list(range(-filtershape//2+1,filtershape//2+1))
    a, b = pixelcoor
    z, w = image2.shape
    neigh_pixs = []
    for col in size:
        for row in size:
            if 0 <= a + row < z and 0 <= b + col < w:
                neigh_pixs.append(image2[a+row,b+col])
    neigh_pixs = np.array(neigh_pixs).astype(np.float)
    if Q>=0:
        out_pixel = np.sum(neigh_pixs**(Q+1)) / (np.sum(neigh_pixs**(Q))+eps)
    else:
        neigh_pixs += eps
        out_pixel = np.sum(neigh_pixs**(Q+1)) / (np.sum(neigh_pixs**(Q))+eps)
    return out_pixel
def contraharmonic_mean_filter(image,filter_size,Q):
    filtered_image = np.zeros(list(image.shape))
    x,y = image.shape
#     count = 0
    for i in range(x):
        for j in range(y):
            filtered_image[i,j] = pix_contraharmonic_mean_filter([i,j], image, filter_size,Q)
    fil_size = filter_size
    return(filtered_image, fil_size)
```

```python
def add_saltpepper_noise_gray(image,salt_prob, pepper_prob):
    row,col = image.shape
    noisy = np.zeros(image.shape)
    for i in range(row):
        for j in range(col):
            noisy[i,j] = np.random.choice([image[i,j],255,0], p=[1-(salt_prob+pepper_prob),salt_prob,pepper_prob])
    np.clip(noisy,0,255)
    return noisy
```

```python
def img_chnnls(rgb):
    channels = []
    ch_count = len(rgb[0][0])
    for i in range(ch_count):
        channels.append(rgb[:,:,i])
    return channels
def dft_chnnls(channelss):
    dfts = []
    for ch in channelss:
        dfts.append((np.fft.fft2(ch)))
    return dfts
```

```python
def athmospheric_turbulence(image, k):
    image_dft_ch = dft_chnnls(img_chnnls(image))
    h = np.zeros(image_dft_ch[0].shape[:2])
    row,col = h.shape
    for i in range(row):
        for j in range(col):
            h[i,j] = e ** (-k*((i**2 + j**2)**(5/6)))
    new_channels = []
    for dft in image_dft_ch:
        new_channels.append(np.clip(np.real(np.fft.ifft2(dft*h)),0,255))
    noisy_image = np.dstack(new_channels)
    noisy_image = noisy_image.astype(np.uint8)
    return noisy_image
```

```python
def motion_blur(image, a, b, t = 5):
    image_dft_ch = dft_chnnls(img_chnnls(image))
    h = np.zeros(image_dft_ch[0].shape[:2]).astype(np.complex)
    row,col = h.shape
    for i in range(row):
        for j in range(col):
            h[i,j] = (e ** (-1j*pi*(i*a + j*b))) *\
            (sin(pi*(i*a + j*b))) *\
            (t/(pi*(i*a + j*b)+eps))
    new_channels = []
    for dft in image_dft_ch:
        new_channels.append(np.clip(np.real(np.fft.ifft2(dft*h)),0,255))
    noisy_image = np.dstack(new_channels)
    noisy_image = noisy_image.astype(np.uint8)
    return noisy_image
```
