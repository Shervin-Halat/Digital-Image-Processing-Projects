```python
main_image1 = cv2.imread('C:/Users/sherw/OneDrive/Desktop/DIP_HW1/inputs/P7/wheres_wally_1.jpg')
main_image2 = cv2.imread('C:/Users/sherw/OneDrive/Desktop/DIP_HW1/inputs/P7/wheres_wally_2.jpg')
main_image3 = cv2.imread('C:/Users/sherw/OneDrive/Desktop/DIP_HW1/inputs/P7/wheres_wally_3.jpg')
main_image4 = cv2.imread('C:/Users/sherw/OneDrive/Desktop/DIP_HW1/inputs/P7/wheres_wally_4.jpg')

color = [0, 0, 0] # 'cause black!

plt.imshow(main_image1)
top, bottom, left, right = [25]*4

img_with_border1 = cv2.copyMakeBorder(main_image1, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)
plt.imshow(img_with_border1)
img_with_border2 = cv2.copyMakeBorder(main_image2, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)
img_with_border3 = cv2.copyMakeBorder(main_image3, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)
img_with_border4 = cv2.copyMakeBorder(main_image4, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)
```

```python
mask1 = np.zeros((img_with_border1.shape[0],img_with_border1.shape[1]))
mask2 = np.zeros((img_with_border2.shape[0],img_with_border2.shape[1]))
mask3 = np.zeros((img_with_border3.shape[0],img_with_border3.shape[1]))
mask4 = np.zeros((img_with_border4.shape[0],img_with_border4.shape[1]))

red_r,red_g,red_b = 237,67,80
white_r,white_g,white_b = 227,238,235

for i in range(20,len(img_with_border1)-20):
    for j in range(20,len(img_with_border1[0])-20):
        if abs(red_r - img_with_border1[i][j][0]) <= 7 and \
        abs(red_g - img_with_border1[i][j][1]) <= 7 and \
        abs(red_b - img_with_border1[i][j][2]) <= 7 or \
        abs(white_r - img_with_border1[i][j][0]) <= 7 and \
        abs(white_g - img_with_border1[i][j][1]) <= 7 and \
        abs(white_b - img_with_border1[i][j][2]) <= 7:

            mask1[i-20:i+21,j-20:j+21] = 1

for i in range(20,len(img_with_border2)-20):
    for j in range(20,len(img_with_border2[0])-20):
        if abs(red_r - img_with_border2[i][j][0]) <= 7 and \
        abs(red_g - img_with_border2[i][j][1]) <= 7 and \
        abs(red_b - img_with_border2[i][j][2]) <= 7 or \
        abs(white_r - img_with_border2[i][j][0]) <= 7 and \
        abs(white_g - img_with_border2[i][j][1]) <= 7 and \
        abs(white_b - img_with_border2[i][j][2]) <= 7:

            mask2[i-20:i+21,j-20:j+21] = 1

for i in range(20,len(img_with_border3)-20):
    for j in range(20,len(img_with_border3[0])-20):
        if abs(red_r - img_with_border3[i][j][0]) <= 7 and \
        abs(red_g - img_with_border3[i][j][1]) <= 7 and \
        abs(red_b - img_with_border3[i][j][2]) <= 7 or \
        abs(white_r - img_with_border3[i][j][0]) <= 7 and \
        abs(white_g - img_with_border3[i][j][1]) <= 7 and \
        abs(white_b - img_with_border3[i][j][2]) <= 7:

            mask3[i-20:i+21,j-20:j+21] = 1
            
for i in range(20,len(img_with_border4)-20):
    for j in range(20,len(img_with_border4[0])-20):
        if abs(red_r - img_with_border4[i][j][0]) <= 7 and \
        abs(red_g - img_with_border4[i][j][1]) <= 7 and \
        abs(red_b - img_with_border4[i][j][2]) <= 7 or \
        abs(white_r - img_with_border4[i][j][0]) <= 7 and \
        abs(white_g - img_with_border4[i][j][1]) <= 7 and \
        abs(white_b - img_with_border4[i][j][2]) <= 7:

            mask4[i-20:i+21,j-20:j+21] = 1         
```

```python
rgbArray1 = np.zeros((img_with_border1.shape[0],img_with_border1.shape[1],3), 'uint8')
rgbArray1[:,:,0] = np.multiply(img_with_border1[:,:,2],mask1)
rgbArray1[:,:,1] = np.multiply(img_with_border1[:,:,1],mask1)
rgbArray1[:,:,2] = np.multiply(img_with_border1[:,:,0],mask1)

rgbArray2 = np.zeros((img_with_border2.shape[0],img_with_border2.shape[1],3), 'uint8')
rgbArray2[:,:,0] = np.multiply(img_with_border2[:,:,2],mask2)
rgbArray2[:,:,1] = np.multiply(img_with_border2[:,:,1],mask2)
rgbArray2[:,:,2] = np.multiply(img_with_border2[:,:,0],mask2)

rgbArray3 = np.zeros((img_with_border3.shape[0],img_with_border3.shape[1],3), 'uint8')
rgbArray3[:,:,0] = np.multiply(img_with_border3[:,:,2],mask3)
rgbArray3[:,:,1] = np.multiply(img_with_border3[:,:,1],mask3)
rgbArray3[:,:,2] = np.multiply(img_with_border3[:,:,0],mask3)

rgbArray4 = np.zeros((img_with_border4.shape[0],img_with_border4.shape[1],3), 'uint8')
rgbArray4[:,:,0] = np.multiply(img_with_border4[:,:,2],mask4)
rgbArray4[:,:,1] = np.multiply(img_with_border4[:,:,1],mask4)
rgbArray4[:,:,2] = np.multiply(img_with_border4[:,:,0],mask4)

# org = cat(np.multiply(img_with_border[:,:0],mask),np.multiply(img_with_border[:,:1],mask),np.multiply(img_with_border[:,:2],mask))
```

```python
def matchTemplate(image,template):

    err = np.sum((image.astype("float") - template.astype("float")) ** 2)
    err /= float(image.shape[0] * image.shape[1])
    return err
```

```python
main_image = cv2.imread('modify_wheres_wally_1.jpeg')
gray_image = cv2.cvtColor(main_image, cv2.COLOR_BGR2GRAY)
template = cv2.imread('C:/Users/sherw/OneDrive/Desktop/DIP_HW1/inputs/P7/wally_5.jpg',0)

mse_matrix = np.zeros((main_image.shape[0],main_image.shape[1])) + 99999
x_axis_leng = len(main_image[0]) - len(template[0])
y_axis_hgt = len(main_image) - len(template)

for i in range(len(gray_image)):
    for j in range(len(gray_image[0])):
        if (i <= y_axis_hgt) and (j <= x_axis_leng):
            if gray_image[i,j] == 0:
                mse_matrix[i,j] = 99999 
            else:
                mse_value = matchTemplate(gray_image[i:i+len(template),j:j+len(template[0])], template)
                mse_matrix[i,j] = mse_value

ind = np.unravel_index(np.argmin(mse_matrix, axis=None), mse_matrix.shape)

bottom_right = (ind[0] + len(template[0]), ind[1] + len(template))
print(bottom_right)
cv2.rectangle(main_image,ind, bottom_right, (220, 0, 50), 6)
plt.imshow(main_image)
plt.show(main_image)
```
